## Individual Season Analysis

```python
import tropycal.tracks as tracks
```

```python
basin = tracks.TrackDataset(basin='north_atlantic',source='hurdat',include_btk=False)
```

```python
season = basin.get_season(2024)
```

```python
print(season)
```

```python
season.to_dataframe()
```

```python
season.summary()
```

```python
season.plot();
```
